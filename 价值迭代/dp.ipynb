## 价值迭代


#### 1、创建环境

```python
def get_state(row, col):
    if row != 3:
        return 'ground'
    if row == 3 and col == 0:
        return 'ground'
    if row == 3 and col == 11:
        return 'terminal'
    
    return 'trap'

get_state(0, 0)
    
```

#### 2、动作空间

```python
def move(row, col ,action):
    if get_state(row, col) in ['trap', 'terminal']:
        return row, col, 0
    
    if action == 0:
        row -= 1

    if action == 1:
        row += 1 

    if action == 2:
        col -= 1 

    if action == 3:
        col += 1

    row = max(0, row) 
    row = min(3, row) 
    col = max(0, col) 
    col = min(11, col) 
    reward = -1;
    if get_state(row, col) == 'trap':
        reward = -100
    if get_state(row, col) == 'terminal':
        reward = 0
    return row, col, reward
```

#### 3、Q函数


```python
import numpy as np
values = np.zeros([4, 12])
pi = np.ones([4, 12, 4]) * 0.25


def get_Qsa(row, col, action):
    next_row, next_col, reward = move(row, col, action)

    value = values[next_row, next_col] * 0.9

    if get_state(next_row, next_col) in ['trap', 'terminal']:
        value = 0
    print("value:",value)
    return value + reward


#策略评估函数
def get_values():
    new_values = np.zeros([4, 12])
    for row in range(4):
        for col in range(12):
            action_value = np.zeros(4)

            for action in range(4):
                action_value[action] = get_Qsa(row, col, action)

            # action_value *= pi[row, col]
            new_values[row, col] = action_value.max()  ## 这里和值迭代和策略迭代的唯一不同，主要体现在，
                                                        ## 值迭代计算完每个状态value后，直接取最大的，而策略迭代则是所有状态的value平均
    return new_values

#策略提升函数
def get_pi():
    new_pi = np.zeros([4, 12 ,4])

    for row in range(4):
        for col in range(12):
            action_value = np.zeros(4)

            for action in range(4):
                action_value[action] = get_Qsa(row, col, action)

            count = (action_value == action_value.max()).sum()

            for action in range(4):
                if action_value[action] == action_value.max():
                    new_pi[row, col, action] = 1 / count
                else:
                    new_pi[row, col, action] = 0
    return new_pi
# get_Qsa(0, 0, 0)
# get_values()
get_pi()
```

#### 4、循环迭代策略评估和策略提升

```python
# for _ in range(10):
for _ in range(100):
    values = get_values()
pi = get_pi()

values, pi
```

##### 5、结果显示

```python
#打印游戏，方便测试
def show(row, col, action):
    graph = [
        '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□',
        '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□', '□',
        '□', '□', '□', '□', '□', '□', '□', '□', '□', '○', '○', '○', '○', '○',
        '○', '○', '○', '○', '○', '❤'
    ]

    action = {0: '↑', 1: '↓', 2: '←', 3: '→'}[action]

    graph[row * 12 + col] = action

    graph = ''.join(graph)

    for i in range(0, 4 * 12, 12):
        print(graph[i:i + 12])

from IPython import display
import time


def test():
    #起点在0,0
    row = 0
    col = 0

    #最多玩N步
    for _ in range(200):

        #选择一个动作
        action = np.random.choice(np.arange(4), size=1, p=pi[row, col])[0]

        #打印这个动作
        display.clear_output(wait=True)
        time.sleep(0.1)
        show(row, col, action)

        #执行动作
        row, col, reward = move(row, col, action)

        #获取当前状态，如果状态是终点或者掉陷阱则终止
        if get_state(row, col) in ['trap', 'terminal']:
            break


test()
```
